```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
s_data.fillna("NULL")
```

```python
s_data.interpolate()
```

```python
s_data.fillna('NULL')
```

```python
input_col = ['btc_market_cap', 'btc_n_transactions', 'btc_miners_revenue' ,'btc_cost_per_transaction', 'btc_difficulty', 'btc_hash_rate', 'btc_cost_per_transaction_percent']
X=s_data[input_col]
Y=s_data['btc_market_price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y)
```

```python
linreg =LinearRegression()
```

```python
linreg.fit(X_train, y_train)
```

```python
y_pred= linreg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.mean_squared_error(y_test, y_pred))
```
