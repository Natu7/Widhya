```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.drop(labels=None, axis=0, index=None, columns= ['Date','Open','High','Low','Close','Volume','Dividend','Split'], level=None, inplace=True)
```

```python
df['HL_PCT'] = (df['Adj_High'] - df['Adj_Low'])/df['Adj_Close']*100.0
```

```python
df['PCT_change'] = (df['Adj_Close'] - df['Adj_Open'])/df['Adj_Open']*100.0
```

```python
df.drop(labels=None, axis=0, index=None, columns= ['Adj_High','Adj_Low','Adj_Open'], level=None, inplace=True)
```

```python
df.fillna(value = -99999)
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
feat_cols = ['Adj_Close','Adj_Volume','HL_PCT','PCT_change']
X = df[feat_cols]
Y = df['PCT_change']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(X_train, Y_train)
```

```python
Y_pred = linreg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.mean_squared_error(Y_test, Y_pred))
```
